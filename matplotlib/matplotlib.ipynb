```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#line graph
x = np.arange(1,4)
y = np.random.randint(6,size = 3)

plt.plot(x,y)
plt.show()
```

```python
#legends,titles and labels
x1 = np.arange(1,4)
y1 = np.random.randint(3,8,size =3 )

x2 = np.arange(1,4)
y2 = np.random.randint(5,10,size =3 )

plt.plot(x1,y1, label = 'First Line')
plt.plot(x2,y2, label = 'Second Line')

plt.xlabel("Number of X")
plt.ylabel("Number of Y")
plt.title("Two Line graph\ncheck it out")
plt.legend()
plt.show()
```

```python
#bar chart and histogram

x = [2,4,6,8,10]
y = [6,7,8,2,4]

x1 = [1,3,5,7,9]
y1 = [7,8,2,4,2]


plt.bar(x,y,label = "Bar Chart 1" , color = "blue")
plt.bar(x1,y1,label = "Bar Chart 2" , color = "r")

plt.xlabel("x")
plt.ylabel("y")
plt.title("bar chart")
plt.legend()
plt.show()
```

```python
#histogram

population_ages  = [22,66,12,35,62,12,112,52,67,23,63,21,95,132,52,31,34,67,87,13,54]
bins = np.arange(0,131,10)
#ids = [x for x in range(len(population_ages))]

#plt.bar(ids,population_ages,label = "bar")
plt.hist(population_ages,bins,histtype="bar",rwidth=0.8)

plt.xlabel("x")
plt.ylabel("y")
plt.title("bar chart")
#plt.legend()
plt.show()
```

```python
#scatter plot
x = np.arange(1,9)
y = np.random.randint(1,9,size = 8)

plt.scatter(x,y , label = 'skat1',color = "b" ,s=200 ,marker = "*")

plt.xlabel("x")
plt.ylabel("y")
plt.title("scatter plot")
plt.legend()
plt.show()
```

```python
#stack plot
days = np.arange(1,6)

sleeping = [7,8,6,11,7]
eating = [2,3,4,3,2]
working = [7,8,7,2,2]
playing = [8,5,7,8,13]

plt.plot([],[],label="sleeping",color="m")
plt.plot([],[],label="eating",color="c")
plt.plot([],[],label="working",color="r")
plt.plot([],[],label="playing",color="k")
plt.stackplot(days,sleeping,eating,working,playing , colors = ['m','c','r','k'])

plt.xlabel("x")
plt.ylabel("y")
plt.title("stack Plot")
plt.legend()
plt.show()
```

```python
#pie chart

slices = [7,2,2,13]
activities = ['sleeping','eating','working','playing']
clr = ['c','m','r','b']

plt.pie(slices,
        labels=activities,
        colors=clr
       ,startangle=90,
        shadow=True,
        explode=(0,0.1,0,0),
       autopct='%1.1f%%')

plt.title("pie chart")
plt.legend()
plt.show()
```
